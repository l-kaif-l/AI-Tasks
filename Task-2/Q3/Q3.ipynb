```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
x = np.linspace( 1,10 )

fx = np.sin(x) + np.random.normal( scale = 0.1, size = len(x) )
ddx_fx = np.poly1d( list(fx) ).deriv()
```

```python

adder = np.array( [ ddx_fx[48] ] )
ddx_fx = np.concatenate([ ddx_fx ,adder ])

np.shape(ddx_fx)
```

```python
fig, graph = pl.subplots( 2,2, figsize = ( 10, 5 ) )

graph[0,1].plot( x, ddx_fx, label = "f'(x)" )
graph[0,0].plot( x, fx, label = "f(x)" )

graph[0,0].set_ylabel("f(x)")
graph[0,1].set_ylabel("f ' (x)")

fig.text( 0.5, 0, "x", rotation = "horizontal",fontsize = 20 )

graph[1,0].plot( x, fx + ddx_fx, label = "f(x) + f '(x)" )
graph[1,1].plot( x, fx - ddx_fx, label = "f(x) - f '(x)" )

graph[1,0].set_ylabel("f(x) + f ' (x)")
graph[1,1].set_ylabel("f(x) - f ' (x)")
```
